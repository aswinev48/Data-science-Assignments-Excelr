##### 6) Given a 'marks' variable, print the grade:
##### - 90 and above: 'A'
##### - 75–89: 'B'
##### - 50–74: 'C'
##### - Below 50: 'Fail'


```python
marks=int(input("Enter the marks:"))
if marks>=90:
    print("A")
elif marks>=75 and marks<90:
    print("B")
elif marks>=50 and marks<75:
    print("C")
else:
    print("Fail")
```

##### 7) Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative

```python
n=int(input("Enter the number:"))
if(n>0):
    if n%2==0:
        print("Positive and Even")
    else:
        print("Positive and Odd")
elif(n<0):
    print("Negative")
else:
    print("Zero")
```

### Section C: Exploring Pandas

##### 11) Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
##### - Use .info(), .describe()
##### - Select 'Name' and 'City' columns
##### - Drop 'City' column
##### - Fill any missing values in 'Salary' column with the mean
##### - Remove any duplicate rows


```python
dict1={'Name':['Aswin','Athul','Abhishek','Aatlee','Abhidev','Gokul','Bharath','Sreerag','Karthik','Krishnan'],'Age':[21,20,23,27,28,25,16,29,18,19],
       'City':['Bengaluru','Madiwala','Lalbagh','HSR','Goa','Agara','Depot','Uduppi','Sarakki','Kengeri'],
       'Salary':[120000,50000,25000,15000,67000,40000,150000,30000,]}
dict1['Salary'].extend([np.nan, np.nan])
```

```python
df1=pd.DataFrame(dict1)
```

```python
df1
```

```python
### info()
df1.info()
```

```python
### describe()
df1.describe()
```

```python
### Selecting 'Name' column
df1.Name
```

```python
### Selecting 'City' column
df1.City
```

```python
### Dropping 'City' Column
df1.drop(columns=["City"],inplace=True)
df1
```

```python
### Chcking missing values
df1.isnull().sum()
```

```python
df1.Salary.mean()
```

```python
df1.fillna({'Salary':df1.Salary.mean()},inplace=True)
```

```python

df1.duplicated().sum()
```

```python
df1.drop_duplicates(ignore_index=True,inplace=True)
```
